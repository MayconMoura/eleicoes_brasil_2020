```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(".//dados//dado_final//dados_totais.csv")
```

```python
data_vereador = data.query("DS_CARGO == 'VEREADOR' and DS_SIT_TOT_TURNO == 'ELEITO POR QP'")
```

```python
perc = [0, 0, 0]
anos = [2012, 2016, 2020]
```

```python
i = 0
for ano in anos:
    tot_eleito = data_vereador.query(f"ANO_ELEICAO == {ano}").shape[0]
    tot_eleito_feminino = data_vereador.query(f"DS_GENERO == 'FEMININO' and ANO_ELEICAO == {ano}").shape[0]
    perc[i] = (tot_eleito_feminino/tot_eleito * 100)
    i+= 1
```

```python
print("Percentual feminino nas c칙maras de vereadores do brasil %.2f"  % (tot_eleito_feminino/tot_eleito * 100), "%")
```

```python
plt.plot(anos, perc)
```
